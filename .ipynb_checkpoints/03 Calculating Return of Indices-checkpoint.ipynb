```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
```

```python
# loading the data from yahoo finance
# tickers = ['^GSPC', '^IXIC', '^GDAXI', '^FTSE']
tickers = ['^GSPC', '^IXIC', '^GDAXI']

df = pd.DataFrame()
for t in tickers:
    df[t] = wb.DataReader(t, data_source='yahoo', start='1997-1-1')['Adj Close']
```

```python
(df/df.iloc[0] * 100).plot(figsize=(12,8))
plt.show()
```
